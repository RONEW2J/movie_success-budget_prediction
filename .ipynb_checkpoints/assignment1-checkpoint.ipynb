Summary of the dataset shows that there are no missing values. But the preview shows that the dataset contains values coded as `?`

```python
df[df == '?'] = np.nan
```

```python
df.isna().sum()
```

```python
fill_values = {
    'workclass': df['workclass'].mode()[0],
    'occupation': df['occupation'].mode()[0],
    'native.country': df['native.country'].mode()[0]
}
df.fillna(fill_values, inplace=True)
```

```python
df.isna().sum()
```

```python
for i in df.columns:
    print(f'"{i}" column is have that unique values')
    print(df[i].unique())
    print(f'And have that count\n')
    print(df[i].value_counts())
    print('\n')
```

```python
# changing the target into 0 and 1 instead of the '<=50K','>50K' to be prebared for analysis
df['income'] = df['income'].replace(['<=50K','>50K'],[0,1])
```

```python
plt.figure(figsize=(10,10))
counter = 1
for i in df.select_dtypes(include=np.number).columns:
    plt.subplot(3,3,counter)
    counter += 1
    sns.boxplot(data = df, x = df[i])
    plt.title(i)
plt.tight_layout()
plt.show()
```

Checking numerical feature distribution

```python
plt.figure(figsize=(10,10))
counter = 1
for i in df.select_dtypes(include=np.number).columns:
    plt.subplot(3,3,counter)
    counter += 1
    sns.histplot(data = df, x = df[i])
    plt.title(i)
plt.tight_layout()
plt.show()
```
